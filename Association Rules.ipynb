#### **INSIGHTS**

**Support:** The proportion of transactions in which the base and add items co-occur. Higher support indicates more frequent occurrence.

**Confidence:** The probability of purchasing the "add" item given the "base" item. Higher confidence suggests stronger associations.

**Lift:** A measure of how much more likely the "add" item is purchased when the "base" item is already purchased compared to random chance. A lift > 1 indicates a positive association.

**1.	What is lift and why is it important in Association rules?**

Lift is a metric used to evaluate the strength of an association rule in data mining. It measures how much more likely two (or more) items are to occur together than they would if they were statistically independent.

Formula for Lift:

For a rule A→B, the lift is calculated as:

**Lift(A→B)=Support(A∩B)/Support(A)×Support(B)**

**--> Why Is Lift Important?**

Lift is crucial because it provides insights into the actual significance of an association rule, beyond mere co-occurrence.

**2.	What is support and Confidence. How do you calculate them?**

--> What is Support?

Support measures the frequency of an itemset occurring in the dataset. It helps identify how often an item or a combination of items appears in transactions.

Formula for Support:

For an itemset A :

**Support(A)=Number of transactions containing A/Total number of transactions**

--> What is Confidence?

Confidence measures the likelihood of occurrence of an item B, given that item A has already occurred. It quantifies the predictive power of an association rule.

Formula for Confidence:

For a rule A→B:

**Confidence(A→B)=Support(A∩B)/Support(A)**


**3.	What are some limitations or challenges of Association rules mining?**

Association rule mining is a powerful technique for discovering relationships between items in a dataset. However, it comes with several limitations and challenges that can affect its applicability and effectiveness. Here are the main ones:
________________________________________
1. Scalability and Performance Issues
•	Large Datasets: As the dataset grows in size (number of transactions or items), the computational complexity of generating frequent itemsets and rules increases significantly.
•	High Dimensionality: Datasets with many items can lead to a combinatorial explosion in the number of possible itemsets, making the process resource-intensive.
________________________________________
2. Handling Sparse Data
•	In datasets where most transactions contain only a few items (e.g., online shopping carts), the frequent itemsets might be too few, and mining meaningful associations can be challenging.
________________________________________
3. Setting Thresholds (Support and Confidence)
•	Support Threshold: If set too high, interesting but less frequent patterns may be missed. If set too low, it may generate too many itemsets, including irrelevant ones.
•	Confidence Threshold: A high confidence threshold might filter out potentially valuable rules, while a low threshold might include unreliable associations.
________________________________________
4. Lack of Contextual Understanding
•	Association rules do not consider the context or causality of the relationships. For instance, a rule A→BA \rightarrow BA→B doesn't imply that AAA causes BBB; it only indicates co-occurrence.
•	Rules may sometimes reflect spurious or coincidental relationships.
________________________________________
5. Difficulty with Continuous or Quantitative Data
•	Association rule mining works best with categorical data. For continuous or numerical attributes, data must be discretized (e.g., converting age into age ranges), which can lead to a loss of information or introduce bias.
________________________________________
6. Handling Null or Missing Data
•	Missing values in datasets can interfere with the identification of frequent patterns. Decisions about whether to exclude, fill, or ignore missing values can impact results.
