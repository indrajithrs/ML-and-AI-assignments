```python
import matplotlib.pyplot as plt
import seaborn as sns

col=df.select_dtypes(include=['float64','int64']).columns
# Loop through each numerical column and plot a Boxplot
fig=plt.figure(figsize=(16, 12))
for i ,column in enumerate(col,1):
    plt.subplot(3,2,i)
    sns.boxplot(x=df[column])
    plt.title(f'Boxplot of{column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    
plt.tight_layout()    
plt.show()
fig.savefig('Boxplot.png')
```

```python

# List of numerical columns
numerical_columns = ['Volume', 'Avg Price', 'Total Sales Value', 'Discount Rate (%)', 'Discount Amount', 'Net Sales Value']

# Loop through each numerical column and create a boxplot
for column in numerical_columns:
    plt.figure(figsize=(8, 6))  # Adjust figure size as needed
    sns.boxplot(x=df[column])
    plt.title(f"Boxplot for {column}")
    plt.show()

    # Identify outliers using the IQR method
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]

    if not outliers.empty:
        print(f"Outliers found in {column}:")
        print(outliers[[column]])
    else:
        print(f"No outliers found in {column}.")
```

```python
import matplotlib.pyplot as plt
import pandas as pd


dis=df.drop(df.columns[[0,4,8,9,10,11,12]],axis=1)
df1= pd.DataFrame(dis)
df1

# Create a figure and axes for the subplots
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(24, 20))
fig.tight_layout(pad=4)  # Adjust the padding between subplots

# List of categorical columns to plot
categorical_columns = ['Day', 'SKU', 'City', 'BU', 'Brand', 'Model']

# Loop through each categorical column and create a bar chart
for i, col in enumerate(categorical_columns):
    row = i // 3
    col_index = i % 3

    # Count the occurrences of each category
    counts = df1[col].value_counts()

    # Create a bar chart on the current axes
    axes[row, col_index].bar(counts.index, counts.values)

    # Set title and labels for the subplot
    axes[row, col_index].set_title(f'Frequency of {col}')
    axes[row, col_index].set_xlabel(col)
    axes[row, col_index].set_ylabel('Count')

    # Rotate x-axis labels for better visibility
    plt.setp(axes[row, col_index].get_xticklabels(), rotation=45, ha='right')

# Display the plots
plt.show()
fig.savefig('Barplot.png')
```
